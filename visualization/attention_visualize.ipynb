```python
attn = torch.load('attention_map.pt')
attn_rec = torch.load('attention_map_rec.pt')
attn2 = torch.load('attention_map2.pt')
words = 'this bird is gray in color, with a yellow belly'.split(' ')
```

```python
def word_heat(attn, words):
    ses.heatmap(attn.sum(-1).sum(-1), cmap="YlGnBu", xticklabels=words)
```

```python
word_heat(attn)
```
